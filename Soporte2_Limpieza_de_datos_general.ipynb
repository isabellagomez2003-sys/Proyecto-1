### Educación de los Padres

```python
df = df.fillna({'FAMI_EDUCACIONMADRE':0, 'FAMI_EDUCACIONPADRE':0})
```

```python
df.isna().sum()
df.notna().sum(axis=0)/df.shape[0]*100
```

```python
df['FAMI_EDUCACIONMADRE'] = df['FAMI_EDUCACIONMADRE'].replace({
    'Ninguno': 1,
    'Primaria incompleta': 2,
    'Primaria completa': 3,
    'Secundaria (Bachillerato) incompleta': 4,
    'Secundaria (Bachillerato) completa': 5,
    'Técnica o tecnológica incompleta': 6,
    'Técnica o tecnológica completa': 7,
    'Educación profesional incompleta': 8,
    'Educación profesional completa': 9,
    'Postgrado': 10,
    'No sabe': 'No Registra',
    'No Aplica': 'No Registra',
    'No Registra': 0
})
```

```python
df['FAMI_EDUCACIONMADRE'].value_counts()
```

```python
df['FAMI_EDUCACIONPADRE'] = df['FAMI_EDUCACIONPADRE'].replace({
    'Ninguno': 1,
    'Primaria incompleta': 2,
    'Primaria completa': 3,
    'Secundaria (Bachillerato) incompleta': 4,
    'Secundaria (Bachillerato) completa': 5,
    'Técnica o tecnológica incompleta': 6,
    'Técnica o tecnológica completa': 7,
    'Educación profesional incompleta': 8,
    'Educación profesional completa': 9,
    'Postgrado': 10,
    'No sabe': 'No Registra',
    'No Aplica': 'No Registra',
    'No Registra': 0
    })
```

```python
df['FAMI_EDUCACIONPADRE'].value_counts()
```

## Limpieza p2

```python
cols_numericas = [
    "PUNT_GLOBAL","PUNT_MATEMATICAS","PUNT_LECTURA_CRITICA",
    "PUNT_C_NATURALES","PUNT_SOCIALES_CIUDADANAS","PUNT_INGLES",
    "FAMI_PERSONASHOGAR","FAMI_CUARTOSHOGAR"
]

for col in cols_numericas:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
#verificando las variables categoricas
df["FAMI_TIENEINTERNET"] = df["FAMI_TIENEINTERNET"].str.strip().str.upper()
df["FAMI_TIENECOMPUTADOR"] = df["FAMI_TIENECOMPUTADOR"].str.strip().str.upper()

#nueva columna de hacimiento (#personas en el hogar / #cuartos)
df["PERSONAS_POR_CUARTO"] = df["FAMI_PERSONASHOGAR"] / df["FAMI_CUARTOSHOGAR"]
```

## FINALIZACIÓN

```python
df.to_csv("df_limpio.csv", index=False)
```
